```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
df1 = pd.read_csv('./Datasets/cervical_cancer/risk_factors_cervical_cancer.csv', na_values='?')
```

```python
df1 = df1.drop(['STDs: Time since first diagnosis', 'STDs: Time since last diagnosis'], axis=1)
```

```python
X = df1.iloc[:,:33]
y = df1.iloc[:, 33]
```

```python
imputer = SimpleImputer(strategy='median')
imputer.fit(X)
X_new = imputer.transform(X)
# y = np.array(y)
```

```python
# y = y.reshape(-1, 1)
```

```python
df_new = pd.DataFrame(X_new, columns = X.columns, index=X.index)
df_new['Biopsy'] = y
```

```python
df_new.to_csv('./Datasets/cervical_cancer/preprocessed_data.csv')
```

```python
df4 = pd.read_csv('./Datasets/banking_dataset/file.csv')

df4
```

```python
df4.columns
```

```python
categorical_features = []
for label in df4.columns:
#     print(label, " ", df4[label].dtype)
    if(df4[label].dtype==object) :
        categorical_features.append(label)
        
print(categorical_features)
```

```python
le = LabelEncoder()

for feature in categorical_features : 
    df4[feature] = le.fit_transform(df4[feature])
```
